```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

def get_product_detail(url):
    """
    get brand, description, 
    """

def crawl_pet_shop_by_url(url):
    # Cấu hình Header để giả lập trình duyệt, tránh bị server chặn

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Kiểm tra nếu link lỗi (404, 500...)
        
        soup = BeautifulSoup(response.content, "html.parser")
        product_list = []

        # Dựa trên HTML bạn gửi, mỗi sản phẩm nằm trong thẻ div class "product-box"
        items = soup.find_all("div", class_="product-box")

        for item in items:
            # 1. Lấy Tên sản phẩm
            # Nằm trong thẻ h3 class "product-name"
            name_tag = item.find("h3", class_="product-name")
            name = name_tag.get_text(strip=True) if name_tag else "Không có tên"

            # 2. Lấy Link chi tiết sản phẩm
            # Lấy từ thuộc tính 'href' của thẻ 'a' bên trong h3
            link_tag = name_tag.find("a") if name_tag else None
            if link_tag and 'href' in link_tag.attrs:
                detail_link = "https://petshophanoi.com" + link_tag['href']
            else:
                detail_link = "N/A"

            # 3. Lấy Giá bán
            # Nằm trong thẻ span class "product-price"
            price_tag = item.find("span", class_="product-price")
            price = price_tag.get_text(strip=True) if price_tag else "Liên hệ"

            product_list.append({
                "Tên sản phẩm": name,
                "Giá bán": price,
                "Link chi tiết": detail_link
            })

        return product_list

    except Exception as e:
        print(f"Đã xảy ra lỗi khi truy cập URL: {e}")
        return []

# --- CHẠY THỬ NGHIỆM ---
target_url = "https://petshophanoi.com/danh-cho-cho"
data = crawl_pet_shop_by_url(target_url)

# Chuyển dữ liệu sang DataFrame để xem dạng bảng
df = pd.DataFrame(data)

# Hiển thị kết quả ra màn hình
print(df.head(10)) # Hiển thị 10 sản phẩm đầu tiên

# Lưu kết quả về file Excel hoặc CSV
# df.to_excel("ket_qua_san_pham.xlsx", index=False)
print(f"\nTổng cộng đã lấy được: {len(df)} sản phẩm.")
```
